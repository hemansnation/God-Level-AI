### write a file

```python
f = open("indore.txt",'w')

f.write("hello from india.")
f.close()
```

```python
f = open("indore.txt",'r')

f.read()
```

### append a file

```python
f = open("student.txt", 'a')

f.write("string from append mode")

f.close()
```

```python
f = open("student.txt", 'r')

print(f.read())
```
