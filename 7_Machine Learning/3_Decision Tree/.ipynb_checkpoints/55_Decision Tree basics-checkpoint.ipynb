# Decision Tree Classification

```python
import pandas as pd
```

```python
df = pd.read_csv('salaries.csv')
df.head()
```

```python
features = df.drop('salary_more_then_100k', axis = 'columns')
features
```

```python
target = df['salary_more_then_100k']
target
```

## Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
features['company_number'] = le.fit_transform(features['company'])
features['job_number'] = le.fit_transform(features['job'])
features['degree_number'] = le.fit_transform(features['degree'])
```

```python
features
```

```python
features = features.drop(['company', 'job', 'degree'], axis='columns')

features
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(features, target)
```

```python
model.score(features, target)
```

```python
model.predict([[2,1,0]])
```

```python
model.predict([[2,1,1]])
```
